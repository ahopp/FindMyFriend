```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib import animation

class Person:
    'Represents a person object in our ficticious world.'
    
    person_count = 0
    
    def __init__(self, name, starting_position, speed=1, memory=30):
        self.name = name
        self.starting_position = starting_position
        self.current_position = self.starting_position
        self.speed = speed
        self.memory = memory
        self.memory_history = [self.current_position]
        self.position_history = [self.current_position]
        Person.person_count += 1
        self.meeting_history = {}
        self.move_adjacency = [(i,j) for i in (-1*self.speed,1*self.speed) for j in range(-1*self.speed,1*self.speed+1,1)] + \
                              [(i,j) for i in range(-1*self.speed+1,1*self.speed,1) for j in (-1*self.speed,1*self.speed)]
        #self.move_adjacency.append([
        #[(i,j) for i in range(-1*self.speed,1*self.speed+1,1) for j in range(-1*self.speed,1*self.speed+1,1) if not (i == j == 0)]
    
    def displayCount(self):
        print "Total People: %d" % Person.person_count
    
    def displayPerson(self):
        print "Name: ", self.name, ", Starting Position: ", self.starting_position
        
    def displayCurrentPosition(self):
        print self.name, ": ", self.current_position
    
    def updatePosition(self, new_position):
        self.current_position = new_position
        self.position_history.append(self.current_position)
        self.memory_history.append(self.current_position)
        if len(self.memory_history) > self.memory:
            self.memory_history = self.memory_history[1::]

    def recordFriendMeet(self, player):
        self.meeting_history.setdefault(player, []).append(self.current_position)
        
    def positionHistoryToList(self):
        return zip(*self.position_history)
    
    def meetingHistoryToList(self,player=None):
        if player in self.meeting_history:
            return zip(*self.meeting_history[player])
        else:
            return [None, None]
        

class Field:
    'Represents our field of play.'
    
    def __init__(self, name, size, fov=1):
        self.name = name
        self.size = size
        self.fov = fov
        self.max_x = size[0]
        self.max_y = size[1]
        self.grid = [[(i,j) for i in range(self.max_x)] for j in range(self.max_y)]
        # Integrate obstacles later.
    
    def displayField(self):
        print "Field name: ", self.name, "; Size: ", self.size, "; Field Of View: ", self.fov

        
class FindMyFriend:
    'Method for implementing the Find My Friend game.'
    
    __game_iterations = 0
    
    def __init__(self, map, players, max_iterations = 500):
        self.map = map
        self.players = players
        self.max_iterations = max_iterations
        self.__game_state = 1
        #self.adjacency = [(i,j) for i in (-1*self.map.fov,0,1*self.map.fov) for j in (-1*self.map.fov,0,1*self.map.fov) if not (i == j == 0)]
        self.collision_adjacency = [(i,j) for i in range(-1*self.map.fov,1*self.map.fov+1,1) for j in range(-1*self.map.fov,1*self.map.fov+1,1)]
        self.__player_dict = {}
        for index, player in enumerate(self.players):
            self.__player_dict[player] = index
        self.__friends_map = np.zeros((len(players), len(players)))
        np.fill_diagonal(self.__friends_map, 1)
        self.__complete_map = np.zeros((len(players), len(players)))
        self.__complete_map[np.tril_indices(len(players))] = 1
        
    def displayMap(self):
        print self.map.displayField()
    
    def displayPlayers(self):
        for player in self.players:
            print player.displayPerson()
    
    def displayPlayerPositions(self):
        for player in self.players:
            print player.displayCurrentPosition()
    
    def getNeighbors(self, player, reason):
        'Reason can be "move" or "collision"'
        neighbors = []
        x_coord = player.current_position[0]
        y_coord = player.current_position[1]
        adjacency = player.move_adjacency if reason=="move" else self.collision_adjacency
        for dx, dy in adjacency:
            if 0 <= (x_coord + dx) < self.map.max_x and 0 <= (y_coord + dy) < self.map.max_y:
                #yield self.map.grid[x_coord + dx][y_coord + dy]
                neighbors.append((x_coord + dx, y_coord + dy))
        return neighbors
    
    def checkPlayerSurroundings(self, player):
        player_neighbors = self.getNeighbors(player, "collision")
        for friend in self.players:
            if friend.current_position in player_neighbors and friend != player:
                self.__friends_map[self.__player_dict[friend]][self.__player_dict[player]] = 1
                self.__friends_map[self.__player_dict[player]][self.__player_dict[friend]] = 1
                player.recordFriendMeet(friend)
                friend.recordFriendMeet(player)
    
    def movePlayer(self, player):
        player_neighbors = self.getNeighbors(player, "move")
        if set(player_neighbors) < set(player.memory_history):
            player.updatePosition(random.choice(player_neighbors))
        else:
            player.updatePosition(random.choice(list(set(player_neighbors) - set(player.memory_history))))
    
    def gameOn(self):
        self.__game_iterations = 0
        for player in self.players:
            self.checkPlayerSurroundings(player)
        while not self.isGameOver():
            for player in self.players:
                self.movePlayer(player)
                self.checkPlayerSurroundings(player)
            self.__game_iterations += 1
        if self.areFriendsFound():
            print "Game finished in %d iterations." % self.__game_iterations
        else:
            print "Game failed to converge in %d iterations." % self.max_iterations
        
    def areFriendsFound(self):
        return (sum(sum(self.__friends_map)) == len(self.__friends_map)**2)
    
    def hasFailedToConverge(self):
        return self.__game_iterations > self.max_iterations
    
    def isGameOver(self):
        return self.areFriendsFound() or self.hasFailedToConverge()
    
    def resetGame(self):
        for player in self.players:
            player.current_position = player.starting_position
            player.meeting_history = {}
            player.position_history = [player.current_position]
            player.memory_history = [player.current_position]
        self.__friends_map = np.zeros((len(self.players), len(self.players)))
        np.fill_diagonal(self.__friends_map, 1)
        self.__game_iterations = 0
    '''
    def prepareAnimation(self, player, init_frame = 0):
        this.animation_x, this.animation_y = player.positionHistoryToList()
        this.animation_iterator = init_frame
    
    def iterateAnimation(self):
        this.animation_iterator += 1
        if this.animation_iterator >= len(this.animation_x):
            this.animation_iterator = 0
        
    def init():
        scat.set_data([], [])
        return scat,

    def animate(i):
        x = this.animation_x[:this.animation_iterator]
        x = this.animation_y[:this.animation_iterator]
        this.iterateAnimation()
        scat.set_data(x, y)
        return scat,
    '''
```
